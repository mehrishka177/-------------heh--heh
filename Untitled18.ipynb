```python
# -----------------------------------
# ğŸ§ Penguin Differentiation Capstone
# -----------------------------------

# ğŸ“¦ Import required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# ğŸ§ª Set up visual style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# ğŸ“ Load the dataset from uploaded CSV
file_path = "/mnt/data/penguins_size.csv"
df = pd.read_csv(file_path)

# ğŸ“ Display first few rows
print("ğŸ“Œ First 5 rows of the dataset:")
print(df.head())

# ğŸ” General info about the dataset
print("\nğŸ“Œ Dataset Information:")
print(df.info())

# ğŸ“Š Summary statistics
print("\nğŸ“Œ Summary Statistics:")
print(df.describe(include='all'))

# ğŸ”§ Check and clean missing values
print("\nğŸ“Œ Missing Values Before Cleaning:")
print(df.isnull().sum())

# Drop missing values
df.dropna(inplace=True)

print("\nğŸ“Œ Missing Values After Cleaning:")
print(df.isnull().sum())

# âœ… Check columns to confirm expected names
print("\nğŸ“Œ Column Names:")
print(df.columns)

# 1ï¸âƒ£ Univariate Analysis
# ------------------------

# ğŸ”¢ Countplot: Species Distribution
sns.countplot(data=df, x="species", palette="Set2")
plt.title("Distribution of Penguin Species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()

# ğŸ“Š Histogram: Bill Length Distribution
sns.histplot(data=df, x="bill_length_mm", hue="species", kde=True, palette="muted")
plt.title("Bill Length Distribution by Species")
plt.xlabel("Bill Length (mm)")
plt.show()

# ğŸ“Š Histogram: Flipper Length Distribution
sns.histplot(data=df, x="flipper_length_mm", hue="species", kde=True, palette="muted")
plt.title("Flipper Length Distribution by Species")
plt.xlabel("Flipper Length (mm)")
plt.show()

# 2ï¸âƒ£ Bivariate Analysis
# ------------------------

# ğŸ” Scatter Plot: Bill Length vs Bill Depth
sns.scatterplot(data=df, x="bill_length_mm", y="bill_depth_mm", hue="species", style="sex", palette="Dark2")
plt.title("Bill Length vs Bill Depth by Species")
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.show()

# ğŸ“¦ Boxplot: Flipper Length by Species
sns.boxplot(data=df, x="species", y="flipper_length_mm", palette="pastel")
plt.title("Flipper Length Distribution Across Species")
plt.xlabel("Species")
plt.ylabel("Flipper Length (mm)")
plt.show()

# ğŸ“¦ Boxplot: Body Mass by Species
sns.boxplot(data=df, x="species", y="body_mass_g", palette="coolwarm")
plt.title("Body Mass Distribution Across Species")
plt.xlabel("Species")
plt.ylabel("Body Mass (g)")
plt.show()

# 3ï¸âƒ£ Pairwise Feature Analysis
# ------------------------------

# ğŸ”„ Pairplot: All numeric features colored by species
sns.pairplot(df, hue="species", palette="colorblind")
plt.suptitle("Pairwise Comparison of Features by Species", y=1.02)
plt.show()

# 4ï¸âƒ£ Group Statistics
# ---------------------

# ğŸ“ˆ Mean values of key numeric features per species
grouped_stats = df.groupby("species")[["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]].mean()
print("\nğŸ“Œ Mean Measurements by Species:")
print(grouped_stats)

# ğŸ“ˆ Mean values split further by sex
grouped_stats_sex = df.groupby(["species", "sex"])[["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]].mean()
print("\nğŸ“Œ Mean Measurements by Species and Sex:")
print(grouped_stats_sex)
```
